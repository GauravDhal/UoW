```python
import numpy as np

rng = np.random.default_rng()

n = 4

m = 2*np.ones((n,n),int)

def detGame(turn , token):
    # turn = who's turn
    # turn = 0 => User's turn // Player 0 
    # turn = 1 => Bot's turn // Player 1
    # token = the player's assigned token
    # token = 0 => Player can play 0
    # token = 1 => Player can play 1
    #ASSUME PLAYER WHO PICKS 0, WIN CON = DET = 0
    global m
    
    # User Case
    if(turn==0):
        print("\nStart of user's turn")
        print(m)
        i,j = userPick()
        print('User picks ',i,j)
        m[i-1,j-1] = token        

    # Bot Case
    elif(turn==1):
        print("\nStart of Bot's Turn")
        print(m)
        i,j = botPick()
        print("Computer picks",i+1,j+1)
        m[i,j] = token

    if(hasTwo(m)):
        detGame( (turn +1) % 2, (token +1) % 2 )
    else:
        print("\n end of game\n",m)
        if(np.linalg.det(m)==0):
            if(token==0):
                print(f"Player {turn} wins")
                return 0
            else:
                print(f"Player {(turn+1)%2} wins")
        else:
            #det = non zero
            if(token==1):
                print(f"Player {turn} wins")
            else:
                print(f"Player {(turn+1)%2} wins")






def userPick():
    print(f"Pick row (1 to {n})")
    i = int(input())
    print(f"pick column (1 to {n})")
    j = int(input())
    #print('Human coordinates',i,j)
    if( 0<i<n+1 and 0<j<n+1 and m[i-1,j-1]==2 ):
        return i, j
    else:
        print("\nInvalid Choice")
        return userPick()

def botPick():
    i = rng.integers(0,n)
    j = rng.integers(0,n)
    if( m[i,j] != 2):
        return botPick()
    return i, j

def hasTwo(m):
    for x in range(0,n):
        for y in range(0,n):
            if( m[x,y] == 2 ):
                return True
    return False
                 


#case 1: User starts, and can play zero
detGame( 0 , 0)

```
